Upon inoculation with 104 TCID50 of virus stock, mean survival times of WT flies were 5.5 days and 3.4 days for the parental stock and the average of all genotypes and replicates of the P10 evolved viruses, respectively. The mean survival time of WT flies decreased from P1 to P5 for viruses evolved in all genotypes (paired t-test, p < 0.005), but from P5 to P10 it only decreased in WT evolved viruses (paired t-test, p = 0.006) (Figure 2D). 


```python
import pandas as pd
from scipy.stats import normaltest
import scipy.stats as stats
from statsmodels.graphics.gofplots import qqplot
import pingouin as pg
```

```python
df = pd.read_csv("../../resources/measurements/20230221_mean_survival_time_comparison.csv", sep=";")
df['mean_survival'] = df['mean_survival'].str.replace(",", ".").astype('float')
df
```

```python
print(pg.normality(df[df["Genotype"]=="WT"], dv='mean_survival', group='Passage'))
```

```python
print(pg.normality(df[df["Genotype"]=="KO"], dv='mean_survival', group='Passage'))
```

```python
print(pg.normality(df[df["Genotype"]=="OE"], dv='mean_survival', group='Passage'))
```

```python
# paired t-test comparing WT-P1 to WT-P5
stats.ttest_ind(df[(df["Genotype"]=="WT") & (df["Passage"]==5)]["mean_survival"],
                df[(df["Genotype"]=="WT") & (df["Passage"]==1)]["mean_survival"],
                alternative="less", equal_var= False).pvalue
```

```python
# paired t-test comparing WT-P1 to WT-P5
stats.ttest_rel(df[(df["Genotype"]=="WT") & (df["Passage"]==5)]["mean_survival"],
                df[(df["Genotype"]=="WT") & (df["Passage"]==1)]["mean_survival"],
                alternative="less", ).pvalue
```

```python
# paired t-test comparing WT-P1 to WT-P5
stats.ttest_rel(df[(df["Genotype"]=="KO") & (df["Passage"]==5)]["mean_survival"],
                df[(df["Genotype"]=="KO") & (df["Passage"]==1)]["mean_survival"],
                alternative="less").pvalue
```

```python
# paired t-test comparing WT-P1 to WT-P5
stats.ttest_rel(df[(df["Genotype"]=="OE") & (df["Passage"]==5)]["mean_survival"],
                df[(df["Genotype"]=="OE") & (df["Passage"]==1)]["mean_survival"],
                alternative="less").pvalue
```
