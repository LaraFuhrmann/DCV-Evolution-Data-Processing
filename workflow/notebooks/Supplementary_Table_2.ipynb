```python
import pandas as pd
```

```python
# load deletions from DCV samples
fname_dcv = '../../resources/deletion_analysis/results/dcv_all_mutations.EB_space.csv'
```

```python
df_dcv = pd.read_csv(fname_dcv)

# add information
df_dcv['n_reads_var'] = df_dcv['Rvar'] + df_dcv['Fvar']
df_dcv['coverage'] = df_dcv['Rtot'] + df_dcv['Ftot']
df_dcv['frequency'] = df_dcv['n_reads_var'] / df_dcv['coverage']

# filtering as done for our samples 
# strand bias test 
df_dcv = df_dcv[df_dcv['Pval']>=0.05]

# minimum read support
minimum_read_support = 10 
df_dcv = df_dcv[df_dcv['n_reads_var']>=minimum_read_support]
```

```python
# check which samples are in that dataframe
print(df_dcv['date'].unique())
print(df_dcv['patient'].unique())
```

```python
# filter for the deletions of interest 
df_dcv = df_dcv[df_dcv["Var"]=='-']
df_dcv_utr = df_dcv[df_dcv['Pos'].isin(list(range(276,285)))]
df_dcv_utr[['Pos', 'Ref', 'Var',  'patient', 'date', 'coverage', 'frequency']]
```
